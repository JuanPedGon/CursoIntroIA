```python
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
```

```python
AllPokemons = [resize(imread("sprites\pokemon\%d.png"%i), (32,32)) for i in range(1,720)]
AllPokemonsFullSize = [imread("sprites\pokemon\%d.png"%i) for i in range(1,720)]
```

```python
import pickle

typedict = pickle.load(open("poke_type_dict", "rb"))
```

```python
typedict[2]
```

```python
X = []
X_ids = []
Y = []

x_size = 30
y_size = 30

index = 1

for i in range(len(AllPokemons)):

        
        pokemon = AllPokemons[index-1]
 
        if not typedict.has_key(index):
            print index, " Not valid"
            
        elif 11 in typedict[index][0:1]:
            print index, " water"
            Y.append(0)
            X.append(pokemon)
            X_ids.append(index)
        elif 10 in typedict[index][0:1]:
            print index, " fire"
            Y.append(1)
            X.append(pokemon)
            X_ids.append(index)
            
        index +=1
```

```python
X = [x[:,:,0:3] for x in X]

X = np.asarray(X)
```

# Enfoque basado en Transfer Learning de CIFAR10

```python
from keras.models import load_model, Model

dcnn = load_model("CIFAR10_trained")

#Retrain PORTATIL
```

```python
dcnn.summary()
```

```python
#Cortamos la red en la capa -4, que se corresponde con "flatten_1 (Flatten)"
#Usamos la red recortada como extractor de características

feat_extractor = Model(dcnn.input, outputs=dcnn.layers[-4].output)

X_feat = feat_extractor.predict(X).reshape((X.shape[0],-1))
print X_feat.shape
```

```python
from sklearn.model_selection import train_test_split

X = np.asarray(X)
Y = np.asarray(Y)
X_ids = np.asarray(X_ids)

# Seleccionamos como pokemon de entrenamiento los de la Primera y segunda generación 

trochipID = np.argmax(X_ids==255)

train_idx = range(X.shape[0])[0:trochipID]
test_idx = range(X.shape[0])[trochipID:]

X_train = X_feat[train_idx]
X_test = X_feat[test_idx]

Y_train = Y[train_idx]
Y_test = Y[test_idx]

X_ids_train = X_ids[train_idx]
X_ids_test = X_ids[test_idx]

print "%d imágenes de entrenamiento y %d de test"%(len(train_idx), len(test_idx))
```

```python
from sklearn.svm import LinearSVC

svc = LinearSVC().fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

print classification_report(Y_test, svc.predict(X_test))
print "\x1b[1mAccuracy: ",svc.score(X_test, Y_test)
```

```python
%matplotlib inline

TYPENAMES = {0:"AGUA", 1:"FUEGO"}

for xid, y, ypred, img in zip(X_ids_test, Y_test, svc.predict(X_test), X[test_idx]):
#   plt.imshow(img)
#   plt.show()

    ax2 = plt.subplot2grid((1, 5), (0, 0), colspan=2)
    ax3 = plt.subplot2grid((1, 5), (0, 3))

    ax2.imshow(AllPokemonsFullSize[xid-1])
    ax3.imshow(img,  interpolation="None")
    
    plt.show()
    
    if y == ypred:
        print " Pokemon num %d, Real: %s, Predicted: %s \n\n"%(xid,TYPENAMES[y],TYPENAMES[ypred])
    else:
        print "\x1b[1m FAIL: Pokemon num %d, Real: %s, Predicted: %s \n\n\x1b[0m"%(xid,TYPENAMES[y],TYPENAMES[ypred])
    
```
